```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Users\EBRU\Documents\machineLearning\week5.task2\insurance.csv",sep=",")

x=df.age.values.reshape(-1,1)
y=df.charges.values.reshape(-1,1)

plt.scatter(x,y)
plt.xlabel("age")
plt.ylabel("charges")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)

y_head= lr.predict(x)

plt.scatter(x,y)
plt.xlabel("age")
plt.ylabel("charges")

plt.plot(x,y_head, color="red", label="linear")
plt.show()
print("10 million person insurance estimation: ", lr.predict([[10000]]))
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_regression= PolynomialFeatures(degree=2)

x_polynomial=polynomial_regression.fit_transform(x)

linear_regression2= LinearRegression()
linear_regression2.fit(x_polynomial,y)

y_head2=linear_regression2.predict(x_polynomial)

plt.scatter(x,y)
plt.xlabel("age")
plt.ylabel("charges")
plt.plot(x,y_head, color="red", label="linear")

plt.plot(x,y_head2, color="green", label="poly")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
print("r_square score for linear regression:", r2_score(y,y_head))
print("r_square score for polynomial regression:", r2_score(y,y_head2))
```
